```python
print("Merging train_transaction and train_identity...")
df = train_transaction.merge(train_identity, on='TransactionID', how='left')

print(f"Original transaction shape: {train_transaction.shape}")
print(f"After merge shape: {df.shape}")
print("Merge complete! From now on we only use df.")
```

```python
# Drop columns where more than 50% of data is missing
threshold = 0.5
missing_pct = df.isnull().sum() / len(df)
cols_to_drop = missing_pct[missing_pct > threshold].index.tolist()

print(f"Columns before dropping: {df.shape[1]}")
print(f"Columns being dropped (>50% missing): {len(cols_to_drop)}")

df = df.drop(columns=cols_to_drop)
print(f"Columns after dropping: {df.shape[1]}")
```

```python
import numpy as np

# Log of transaction amount (reduces skew)
df['TransactionAmt_log'] = np.log1p(df['TransactionAmt'])

# Transaction amount decimals (fraudsters often use round numbers)
df['TransactionAmt_decimal'] = df['TransactionAmt'] - df['TransactionAmt'].astype(int)

# Hour and day already created in EDA — check and recreate cleanly
df['hour'] = (df['TransactionDT'] / 3600 % 24).astype(int)
df['day']  = (df['TransactionDT'] / (3600 * 24) % 7).astype(int)

# Is it a night transaction? (midnight to 6am = higher fraud risk)
df['is_night'] = ((df['hour'] >= 0) & (df['hour'] <= 6)).astype(int)

print("New features created:")
print("  TransactionAmt_log — log scale of amount")
print("  TransactionAmt_decimal — decimal part of amount")
print("  hour — hour of day (0-23)")
print("  day — day of week (0-6)")
print("  is_night — 1 if transaction between midnight and 6am")
```

```python
# These are text columns - we need to convert them to numbers
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f"Categorical columns found: {len(categorical_cols)}")
print(categorical_cols)

# Label encode each one
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in categorical_cols:
    df[col] = df[col].astype(str)  # fill NaN as string first
    df[col] = le.fit_transform(df[col])

print("\nAll categorical columns converted to numbers.")
```

```python
# Fill all remaining NaN with -999 
# XGBoost handles this well and -999 signals "missing" to the model
df = df.fillna(-999)

print(f"Missing values remaining: {df.isnull().sum().sum()}")
print("All missing values filled with -999")
```

```python
# Target = what we are predicting
y = df['isFraud']

# Features = everything else except ID and target
X = df.drop(columns=['isFraud', 'TransactionID'])

print(f"Features shape (X): {X.shape}")
print(f"Target shape (y): {y.shape}")
print(f"\nFraud cases: {y.sum():,} ({y.mean()*100:.2f}%)")
print(f"Non-fraud cases: {(y==0).sum():,} ({(1-y.mean())*100:.2f}%)")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,       # 80% train, 20% test
    random_state=42,     # so results are reproducible
    stratify=y           # keeps same fraud ratio in both splits
)

print(f"Training set: {X_train.shape[0]:,} rows")
print(f"Testing set:  {X_test.shape[0]:,} rows")
print(f"\nFraud in training: {y_train.sum():,} ({y_train.mean()*100:.2f}%)")
print(f"Fraud in testing:  {y_test.sum():,} ({y_test.mean()*100:.2f}%)")
print("\nStage 2 Complete! Ready for Stage 3 — XGBoost Model")
```

```python
from xgboost import XGBClassifier

print("Training XGBoost model... this will take 5-10 minutes, please wait.")

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    scale_pos_weight=27,  # handles imbalance (569877/20663 ≈ 27)
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='auc',
    verbosity=1
)

model.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=100
)

print("\nModel training complete!")
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Predictions
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# ROC AUC Score
auc_score = roc_auc_score(y_test, y_pred_proba)
print("=" * 50)
print("MODEL PERFORMANCE")
print("=" * 50)
print(f"ROC AUC Score: {auc_score:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Not Fraud', 'Fraud']))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Fraud', 'Fraud'],
            yticklabels=['Not Fraud', 'Fraud'])
plt.title('Confusion Matrix', fontsize=14)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

print(f"\nTrue Negatives  (correctly said NOT fraud): {cm[0][0]:,}")
print(f"False Positives (wrongly said fraud):       {cm[0][1]:,}")
print(f"False Negatives (missed actual fraud):      {cm[1][0]:,}")
print(f"True Positives  (correctly caught fraud):   {cm[1][1]:,}")
```

```python
import pandas as pd

feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

plt.figure(figsize=(12, 8))
plt.barh(
    feature_importance['feature'].head(20)[::-1],
    feature_importance['importance'].head(20)[::-1],
    color='steelblue'
)
plt.title('Top 20 Most Important Features', fontsize=14)
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

print("\nTop 10 features:")
print(feature_importance.head(10))
```

```python
import pickle

pickle.dump(model, open('fraud_model.pkl', 'wb'))
pickle.dump(list(X_train.columns), open('feature_names.pkl', 'wb'))

print("Model saved as fraud_model.pkl")
print("Feature names saved as feature_names.pkl")
print("\nStage 3 Complete! Ready for Stage 4 — Isolation Forest")
```

```python
from sklearn.ensemble import IsolationForest

print("Training Isolation Forest... please wait.")

iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.035,  # we know fraud rate is 3.5%
    random_state=42,
    n_jobs=-1
)

iso_forest.fit(X_train)

print("Isolation Forest training complete!")
```

```python
# Get anomaly scores for test set
# -1 = anomaly (suspicious), 1 = normal
iso_predictions = iso_forest.predict(X_test)
iso_scores = iso_forest.decision_function(X_test)

# Convert: -1 becomes 1 (fraud), 1 becomes 0 (normal)
iso_fraud_flag = (iso_predictions == -1).astype(int)

print(f"Transactions flagged as anomalies: {iso_fraud_flag.sum():,}")
print(f"Anomaly percentage: {iso_fraud_flag.mean()*100:.2f}%")
```

```python
from sklearn.metrics import classification_report, roc_auc_score

print("=" * 50)
print("ISOLATION FOREST PERFORMANCE")
print("=" * 50)
print(classification_report(y_test, iso_fraud_flag, target_names=['Not Fraud', 'Fraud']))

iso_auc = roc_auc_score(y_test, -iso_scores)
print(f"ROC AUC Score: {iso_auc:.4f}")
```

```python
import pandas as pd

comparison = pd.DataFrame({
    'Model': ['XGBoost (Layer 1)', 'Isolation Forest (Layer 2)'],
    'ROC AUC': [
        round(roc_auc_score(y_test, y_pred_proba), 4),
        round(iso_auc, 4)
    ],
    'Fraud Caught': [
        (y_pred[y_test == 1] == 1).sum(),
        (iso_fraud_flag[y_test == 1] == 1).sum()
    ],
    'Total Fraud in Test': [4133, 4133]
})

print(comparison.to_string(index=False))
print("\nXGBoost catches known fraud patterns.")
print("Isolation Forest catches unusual transactions XGBoost might miss.")
```

```python
import numpy as np

# A transaction is flagged if EITHER model catches it
combined_flag = ((y_pred == 1) | (iso_fraud_flag == 1)).astype(int)

print("=" * 50)
print("COMBINED TWO-LAYER SYSTEM PERFORMANCE")
print("=" * 50)
print(classification_report(y_test, combined_flag, target_names=['Not Fraud', 'Fraud']))

total_fraud = y_test.sum()
caught_combined = combined_flag[y_test == 1].sum()
print(f"Total fraud in test set: {total_fraud:,}")
print(f"Caught by combined system: {caught_combined:,}")
print(f"Fraud detection rate: {caught_combined/total_fraud*100:.2f}%")
```

```python
import shap
import pickle

print("Creating SHAP explainer... this may take 2-3 minutes.")

explainer = shap.TreeExplainer(model)
pickle.dump(explainer, open('explainer.pkl', 'wb'))

print("SHAP explainer created and saved!")
```

```python
# Use 500 samples — full dataset takes too long
sample_size = 500
X_sample = X_test.iloc[:sample_size]
y_sample = y_test.iloc[:sample_size]

print(f"Calculating SHAP values for {sample_size} transactions...")
shap_values = explainer.shap_values(X_sample)
print("SHAP values calculated!")
print(f"Shape of SHAP values: {shap_values.shape}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
shap.summary_plot(
    shap_values, 
    X_sample, 
    plot_type="bar",
    max_display=15,
    show=False
)
plt.title("Top 15 Features Driving Fraud Predictions", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
shap.summary_plot(
    shap_values, 
    X_sample,
    max_display=15,
    show=False
)
plt.title("How Features Push Predictions Towards Fraud or Not Fraud", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Find a fraud transaction in our sample
fraud_indices = y_sample[y_sample == 1].index
first_fraud_idx = fraud_indices[0]
fraud_position = X_sample.index.get_loc(first_fraud_idx)

print(f"Explaining transaction at position {fraud_position}")
print(f"Actual label: FRAUD ✓")
print(f"Model probability: {model.predict_proba(X_sample.iloc[[fraud_position]])[:,1][0]:.2%}")

shap.force_plot(
    explainer.expected_value,
    shap_values[fraud_position],
    X_sample.iloc[fraud_position],
    matplotlib=True,
    show=False
)
plt.title("Why this transaction was flagged as FRAUD", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Find a normal transaction
normal_indices = y_sample[y_sample == 0].index
first_normal_idx = normal_indices[0]
normal_position = X_sample.index.get_loc(first_normal_idx)

print(f"Explaining transaction at position {normal_position}")
print(f"Actual label: NOT FRAUD ✓")
print(f"Model probability: {model.predict_proba(X_sample.iloc[[normal_position]])[:,1][0]:.2%}")

shap.force_plot(
    explainer.expected_value,
    shap_values[normal_position],
    X_sample.iloc[normal_position],
    matplotlib=True,
    show=False
)
plt.title("Why this transaction was NOT flagged as fraud", fontsize=12)
plt.tight_layout()
plt.show()
```
